```python
data['txn_dt'] = data['timestamp'].dt.date
```

Переведем сумму в usd.

```python
data_2 = data.merge(data_exch, how = 'inner', left_on = 'txn_dt', right_on = 'date')
```

```python
data_2.head()
```

```python
# Переводим в USD
data_2['rate'] = [
                x[1] if x[0] == 'AUD' else x[2] if x[0] == 'BRL' else
                x[3] if x[0] == 'CAD' else x[4] if x[0] == 'EUR' else
                x[5] if x[0] == 'GBP' else x[6] if x[0] == 'JPY' else
                x[7] if x[0] == 'MXN' else x[8] if x[0] == 'NGN' else
                x[9] if x[0] == 'RUB' else x[10] if x[0] == 'SGD' else x[11]
                 for x in zip(data_2['currency'], data_2['AUD'], data_2['BRL'],
                              data_2['CAD'], data_2['EUR'], data_2['GBP'], data_2['JPY'],
                              data_2['MXN'], data_2['NGN'], data_2['RUB'], data_2['SGD'],
                              data_2['USD']
                 )
]
```

```python
data_2['amount_usd'] = data_2['amount']/data_2['rate']
```

```python
int(data_2[data_2['is_fraud'] == 0]['amount_usd'].mean())
```

```python
int(data_2[data_2['is_fraud'] == 1]['amount_usd'].mean())
```

Средняя сумма fraud тразакций выше, чем не fraud (874 против 459 usd).

```python
data_2['unique_merchants_last_hour'] = data_2['last_hour_activity'].apply(lambda x: x.get('unique_merchants'))
```

```python
data_2_clnt = data_2.groupby('customer_id', as_index=False)['unique_merchants_last_hour'].median()
```

```python
data_2_clnt['unique_merchants_last_hour'].quantile(0.95)
```

Посмотрим, с какими признаками коррелирует целевая переменная.

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['is_fraud', 'is_card_present', 'is_outside_home_country', 'is_high_risk_vendor', 'is_weekend',
        'txn_hour', 'amount_usd', 'unique_merchants_last_hour']

corr_matrix = data_2[cols].corr()
plt.figure(figsize=(8, 6)) # Adjust figure size as needed
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

Видим, что доля fraud транзакций выше при оплате за пределами страны клиента и когда карта присутствовала при оплате.
